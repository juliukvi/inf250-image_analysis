```python
from skimage.viewer import ImageViewer
import matplotlib.pyplot as plt
from skimage import io
```

```python
# reading image
filename = "fall.tif"
fall = io.imread(filename)
```

```python
# display image
plt.imshow(fall)
```

```python
# shape of image array
fall.shape
```

```python
# extracting one color and finding min and max values
fall_red = fall[:, :, 0]
fall_green = fall[:, :, 1]
fall_blue = fall[:, :, 2]
print(f"Max red value: {fall_red.flatten().max()}")
print(f"Max green value: {fall_green.flatten().max()}")
print(f"Max blue value: {fall_blue.flatten().max()}")
```

```python
# plot a single color 
plt.imshow(fall_red, vmin=0, vmax=255)
plt.show()
```

```python
# another viewer
viewer = ImageViewer(fall)
viewer.show()
```
