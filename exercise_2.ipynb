```python
import skimage.io
import matplotlib.pyplot as plt
import numpy as np
import skimage.filters
```

```python
# read the image
airfield = skimage.io.imread("images/airfield.tif")
```

```python
# create historgram object
airfield_hist = skimage.exposure.histogram(airfield)
```

```python
# show image
skimage.io.imshow(airfield)
```

```python
# show historgram
plt.plot(airfield_hist[1], airfield_hist[0])
```

```python
# carry out equalisation
airfield_equal = skimage.exposure.equalize_hist(airfield)
```

```python
# show equalised image
skimage.io.imshow(airfield_equal)
```

```python
# create histogram of equalised image
airfield_equal_hist = skimage.exposure.histogram(airfield_equal)
```

```python
# plot equalized image histogram
plt.plot(airfield_equal_hist[1], airfield_equal_hist[0])
```

```python
# do outsu thresholding and show image
airfield_tresh = skimage.filters.threshold_otsu(airfield)
binary = airfield > airfield_tresh
skimage.io.imshow(binary)
```
